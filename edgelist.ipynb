```python
import re
import itertools
path = "data/"
#filenames = ["T","S","TS","Tweet"]
filenames = ["Tweet"]
window_size = 5

for filename in filenames:
    dic = {}
    wordnum = 0
    storefile = filename+"_edgelist.txt"
    with open(path+filename,'r')as datafile:
        sf = open(storefile,'w')
        sents = datafile.readlines()
        punc = '",:'
        for data in sents:
            data = re.sub(r'[{}]+'.format(punc),'',data)
            data = data.split(' ')[1:-2]
            plist = []
            for word in data:
                if len(word) > 0:
                    word = word.lower()
                    #print(word)
                    if word not in dic:
                        #print(wordnum, word)
                        dic[word] = wordnum
                        wordnum += 1
                    plist.append(dic[word])
                    
            for l in range(len(plist)):
                for m in range(l-window_size, l+window_size+1):
                    if m<0 or m>=len(plist): continue
                    sf.write(str(plist[l])+" "+str(plist[m])+'\n')
            
    sf.close()
    datafile.close()
    
    dicf = open(filename+"_dictionary.txt",'w')
    for dic_key in dic.keys():
        dicf.write(dic_key+" "+str(dic[dic_key])+'\n')
        #print(dic_key," ",str(dic[dic_key]))
    dicf.close()
```
